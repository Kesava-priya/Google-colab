**1.** Write a Pandas program to count number of columns of a DataFrame. Sample Output:
Original DataFrame
col1 col2 col3
0 1 4 7
1 2 5 8
2 3 6 12
3 4 9 1
4 7 5 11
Number of columns:
3


```python
import pandas as pd

# Create the original DataFrame
data = {'col1': [1, 2, 3, 4, 7],
        'col2': [4, 5, 6, 9, 5],
        'col3': [7, 8, 12, 1, 11]}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Count the number of columns
num_columns = df.shape[1]

# Display the number of columns
print("\nNumber of columns:", num_columns)
```

**2.** Use the diabetes data set from UCI data set for performing the following:
Apply Univariate analysis:  Frequency
 Mean,  Median,  Mode,  Variance  Standard Deviation
 Skewness and Kurtosis

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Load the Pima Indians Diabetes dataset
data = pd.read_csv('/content/drive/MyDrive/diabetes.csv')

# Univariate Analysis

# Frequency
frequency_outcome = data['Outcome'].value_counts()
print("Frequency of Outcome (Diabetes):")
print(frequency_outcome)

# Mean
mean_values = data.mean()
print("\nMean values for each variable:")
print(mean_values)

# Median
median_values = data.median()
print("\nMedian values for each variable:")
print(median_values)

# Mode
mode_values = data.mode().iloc[0]
print("\nMode values for each variable:")
print(mode_values)

# Variance
variance_values = data.var()
print("\nVariance values for each variable:")
print(variance_values)

# Standard Deviation
std_deviation_values = data.std()
print("\nStandard Deviation values for each variable:")
print(std_deviation_values)

# Skewness
skewness_values = data.apply(skew)
print("\nSkewness values for each variable:")
print(skewness_values)

# Kurtosis
kurtosis_values = data.apply(kurtosis)
print("\nKurtosis values for each variable:")
print(kurtosis_values)
```

**3.** Compare the results of the Univariate and Bivariate analysis for the UCI diabetes dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('/content/drive/MyDrive/diabetes.csv')

# Univariate Analysis
sns.histplot(data['Age'], kde=True)
plt.title('Distribution of Age')
plt.show()

# Bivariate Analysis
sns.scatterplot(x='Age', y='BMI', hue='Outcome', data=data)
plt.title('Scatter Plot of Age vs. BMI with Outcome')
plt.show()
```

5. Write a Pandas program to check whether a given column is present in a DataFrameornot. Sample data:
Original DataFrame
col1 col2 col3
0 1 4 7
1 2 5 8
2 3 6 12
3 4 9 1
4 7 5 11
Col4 is not present in DataFrame. Col1 is present in DataFrame

```python
import pandas as pd

# Create the original DataFrame
data = {'col1': [1, 2, 3, 4, 7],
        'col2': [4, 5, 6, 9, 5],
        'col3': [7, 8, 12, 1, 11]}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Check if 'Col4' is present in DataFrame
col_to_check = 'Col4'
if col_to_check in df.columns:
    print(f"\n{col_to_check} is present in DataFrame.")
else:
    print(f"\n{col_to_check} is not present in DataFrame.")

# Check if 'Col1' is present in DataFrame
col_to_check = 'col1'
if col_to_check in df.columns:
    print(f"{col_to_check} is present in DataFrame.")
else:
    print(f"{col_to_check} is not present in DataFrame.")
```

6 i. Write a NumPy program to perform the basic arithmetic operations


```python
import numpy as np

# Create two NumPy arrays
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])

# Perform basic arithmetic operations
addition_result = np.add(array1, array2)
subtraction_result = np.subtract(array1, array2)
multiplication_result = np.multiply(array1, array2)
division_result = np.divide(array1, array2)

# Display the results
print("Array 1:")
print(array1)

print("\nArray 2:")
print(array2)

print("\nAddition Result:")
print(addition_result)

print("\nSubtraction Result:")
print(subtraction_result)

print("\nMultiplication Result:")
print(multiplication_result)

print("\nDivision Result:")
print(division_result)
```

6 ii.Write a NumPy program to transpose an array

```python
import numpy as np

# Create a NumPy array
original_array = np.array([[1, 2, 3], [4, 5, 6]])

# Transpose the array
transposed_array = np.transpose(original_array)

# Display the original and transposed arrays
print("Original Array:")
print(original_array)

print("\nTransposed Array:")
print(transposed_array)
```
